```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Gender'].value_counts().plot(kind='bar')
plt.title("Customer Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Number of Customers")
plt.show()
```

```python
df['Category'].value_counts().plot(kind='bar')
plt.title("Products Purchased by Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
# Age Distribution

plt.figure(figsize=(8,5))

plt.hist(df['Age'], bins=10)

plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Customers")

plt.show()
```

### Insight

- The chart shows how customers are distributed across different age groups.
- It helps identify the dominant customer age segment.
- Businesses can use this information to create targeted marketing campaigns.

```python
# Purchase Amount Distribution

plt.figure(figsize=(8,5))

plt.hist(df['Purchase Amount (USD)'], bins=10)

plt.title("Purchase Amount Distribution")
plt.xlabel("Purchase Amount (USD)")
plt.ylabel("Number of Customers")

plt.show()
```

### Insight

- The chart shows the distribution of customer purchase amounts.
- It helps identify common spending patterns.
- Businesses can use this information to design pricing and promotional strategies.

```python
# Review Rating Distribution

plt.figure(figsize=(8,5))

plt.hist(df['Review Rating'], bins=10)

plt.title("Review Rating Distribution")
plt.xlabel("Review Rating")
plt.ylabel("Number of Customers")

plt.show()
```

### Insight

- The chart shows how customer review ratings are distributed.
- It helps evaluate customer satisfaction levels.
- Businesses can identify opportunities to improve customer experience.

```python
# Top 10 Locations by Revenue

location_sales = df.groupby('Location')['Purchase Amount (USD)'].sum().sort_values(ascending=False).head(10)

location_sales.plot(kind='bar', figsize=(10,5))

plt.title("Top 10 Locations by Revenue")
plt.xlabel("Location")
plt.ylabel("Revenue (USD)")

plt.show()
```

### Insight

- The chart identifies the locations generating the highest revenue.
- These locations contribute significantly to overall business sales.
- Businesses can prioritize marketing and inventory in these regions.

```python
# Sales by Category

category_sales = df.groupby('Category')['Purchase Amount (USD)'].sum()

category_sales.plot(kind='bar', figsize=(8,5))

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Revenue (USD)")

plt.show()
```

### Insight

- The chart compares revenue across product categories.
- It helps identify the highest-performing product categories.
- Businesses can focus on promoting high-revenue categories.

```python
# Payment Method Preference

df['Payment Method'].value_counts().plot(kind='bar', figsize=(8,5))

plt.title("Preferred Payment Methods")
plt.xlabel("Payment Method")
plt.ylabel("Number of Customers")

plt.show()
```

### Insight

- The chart shows the payment methods preferred by customers.
- It helps businesses understand customer payment preferences.
- Businesses can improve the checkout experience by supporting popular payment methods.
